# Shipwreck Survival Rate Project,2020

### This project aims at predicting the survival rate of a passenger on a voyage based on the Pclass(Passanger class), sex, age and fare. Random Forest,a tool in machine learning was used to design the model. Accordingly, predictions were made and found to be correct from the accuracy score, other things being equal.

Import libraries and dataset

```python
#import libraries for data exploration and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
df = pd.read_csv('titanic.csv')
df.head(10)
```

```python
df.tail(10)
```

```python
#check data for inconsistencies
df.apply(lambda x: sum(x.isna()), axis=0)
```

We need to clean the Age column since we need it for making our model

```python
df['Age'].fillna(int(df['Age'].mean()), inplace=True)
```

```python
#check the correlation of the independent column with the dependent variable(Survived)
#sn.pairplot(df, x_vars=['Pclass','Age','Fare'], y_vars='Survived', height=7, aspect=1, kind='Reg')
```

```python
#change the Sex column to integer
df_dummy = pd.get_dummies(df['Sex'])
df_dummy.head()
```

```python
#concatenate the two tables
X = pd.concat([df[['Pclass','Age','Fare']], df_dummy], axis='columns')
X.head()
```

```python
y = df['Survived']
```

### Machine Learning

```python
#import libraries for machine learning
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#import the random forest class
from sklearn.ensemble import RandomForestRegressor
```

```python
#instantiate the decision tree model
model = RandomForestRegressor()
```

```python
#train the model
model.fit(X_train, y_train)
```

```python
#check the accuracy of the model
model.score(X_test,y_test)
```

### Predictions

Criteria 1
a. Pclass is 2
b. Age is 30
c. Sex is male
d. Fare is 55.00
What is the survival state?

```python
y_pred = model.predict([[2, 30.0, 0, 1, 55.00]])
y_pred
```

Someone with the above criteria will survive

```python
y_pred = model.predict([[1, 15.0, 1, 0, 15.00]])
y_pred
```

Someone with the above criteria will most likely not survive if there was a shipwreck. This is naturally true as someone at age 15 is most likely to be inexperienced on sea.
